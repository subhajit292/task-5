```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("train.csv")
print("Dataset Info:")
print(df.info())
print("\nDataset Description:")
print(df.describe(include='all'))

print("\nMissing Values:")
print(df.isnull().sum())
print("\nSex Value Counts:")
print(df['Sex'].value_counts())
print("\nEmbarked Value Counts:")
print(df['Embarked'].value_counts())
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols].hist(bins=20, figsize=(15, 10))
plt.suptitle("Histograms of Numerical Features")
plt.show()
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age Distribution by Survival")
plt.show()
plt.figure(figsize=(10,8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
selected_cols = ['Survived', 'Pclass', 'Age', 'Fare']
sns.pairplot(df[selected_cols].dropna(), hue='Survived', diag_kind='hist')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
print("\n--- Observations ---")
print("1. Majority of passengers are male, but females have higher survival rates.")
print("2. Higher passenger class (Pclass 1) has higher survival rates.")
print("3. There are missing values in Age, Cabin, and Embarked.")
print("4. Fare and Pclass are negatively correlated (-0.55).")
print("5. Young children had higher chances of survival.")

```
